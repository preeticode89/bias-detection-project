```python
import pandas as pd

data = pd.DataFrame({
    "age": [25, 45, 35, 50, 23, 40],
    "workclass": ["Private", "Self", "Private", "Gov", "Private", "Self"],
    "education": ["Bachelors", "Masters", "PhD", "Bachelors", "HS-grad", "Masters"],
    "sex": ["Male", "Female", "Male", "Female", "Male", "Female"],
    "hours-per-week": [40, 50, 45, 60, 30, 55],
    "income": [0, 1, 1, 0, 0, 1]
})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# features aur target alag
X = data.drop('income', axis=1)
y = data['income']

# text ko number me convert
X = pd.get_dummies(X)

# split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# model train
model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Model ready ")
```

```python
from fairlearn.metrics import MetricFrame, selection_rate

# sensitive feature (gender)
sensitive_feature = X_test['sex_Male']

# bias metrics
metrics = {
    "selection_rate": selection_rate
}

metric_frame = MetricFrame(
    metrics=metrics,
    y_true=y_test,
    y_pred=model.predict(X_test),
    sensitive_features=sensitive_feature
)

print(metric_frame.by_group)
```

```python
from fairlearn.reductions import ExponentiatedGradient, DemographicParity
from sklearn.ensemble import RandomForestClassifier

# naya fair model
mitigator = ExponentiatedGradient(
    RandomForestClassifier(),
    constraints=DemographicParity()
)

# train with fairness
mitigator.fit(X_train, y_train, sensitive_features=X_train['sex_Male'])

# new predictions
y_pred_fair = mitigator.predict(X_test)

print("Fair model ready ")
```

```python
from fairlearn.metrics import MetricFrame, selection_rate

# BEFORE (old model)
metric_before = MetricFrame(
    metrics={"selection_rate": selection_rate},
    y_true=y_test,
    y_pred=model.predict(X_test),
    sensitive_features=X_test['sex_Male']
)

# AFTER (fair model)
metric_after = MetricFrame(
    metrics={"selection_rate": selection_rate},
    y_true=y_test,
    y_pred=y_pred_fair,
    sensitive_features=X_test['sex_Male']
)

print("Before mitigation:\n", metric_before.by_group)
print("\nAfter mitigation:\n", metric_after.by_group)
```
